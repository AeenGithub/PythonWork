```python
# coding=utf-8
import urllib,urllib.request,sys,json
import pandas as pd
import os
import socket
import hashlib
import time
# timeout in seconds
from functools import reduce
import urllib,urllib.request,sys,json
import pandas as pd
import os
import socket
import hashlib
import time
from lxml import etree
from lxml.etree import HTMLParser
# timeout in seconds
from functools import reduce
import requests
timeout = 5
socket.setdefaulttimeout(timeout)


def stock_data(url):
    try:
        headers = { 
                            'Connection':'keep-alive',
                            'Cache-Control':'max-age=0',
                            'Accept': 'text/html, */*; q=0.01',
                            'X-Requested-With': 'XMLHttpRequest',
                            'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2272.89 Safari/537.36',
                            'DNT':'1',
                            'Referer': 'http://example.com/',
                            'Accept-Encoding': 'gzip, deflate, sdch',
                            'Accept-Language': 'zh-CN,zh;q=0.8,ja;q=0.6'
                  }
#         request = urllib.request.Request(url, headers) #15days
#         response = urllib.request.urlopen(request)
#         content = response.read()
        res = requests.get(url,headers= headers)
        #分割数据，给city赋值的同时返回data。
        json_data = json_text = etree.tostring(etree.HTML(res.text,etree.HTMLParser()),encoding="unicode").replace("<html><body><p>","").replace("</p></body></html>","")
        return json_data
    except Exception as e:
        print (e)
    

def write_json(data,path):
    with open(path,'w+') as f:
        f.write(str(data))
        print(path)
        print('writeing')

def read_json(path):
    with open(path,'r',encoding='gbk') as f:
        tem = f.read()
        print(path)
    return tem


def parse_json(path):
    try:
        area = [_.get("area") for _ in json.loads((read_json(path))).get('listItem')]
        detail = [_.get("medal").get("ajaxurl") for _ in json.loads((read_json(path))).get('listItem')]
        favRate = [_.get("favRate") for _ in json.loads((read_json(path))).get('listItem')]
        name = [_.get("shop").get("name") for _ in json.loads((read_json(path))).get('listItem')]
        logo = [_.get("shop").get("logo") for _ in json.loads((read_json(path))).get('listItem')]
        keywords = [_.get("shop").get("keyBiz") for _ in json.loads((read_json(path))).get('listItem')]
        ids = [_.get("shop").get("id") for _ in json.loads((read_json(path))).get('listItem')]
        isMall = [_.get("shop").get("isMall") for _ in json.loads((read_json(path))).get('listItem')]
        urls = [_.get("shop").get("url") for _ in json.loads((read_json(path))).get('listItem')]
        totalSold = [_.get("shop").get("totalSold") for _ in json.loads((read_json(path))).get('listItem')]
        pd_tem = pd.DataFrame({"area":area,"detail":detail,"favRate":favRate,"name":name,"logo":logo,"keywords":keywords,"ids":ids,
                     "ids":ids,"isMall":isMall,"urls":urls,"totalSold":totalSold})
    except:
        pd_tem = pd.DataFrame(columns=['area','detail','favRate','name','logo','keywords','ids','isMall','urls','totalSold'])
    return pd_tem
```
