```python
import pandas as pd
from fake_useragent import UserAgent
import requests,json
from lxml import etree
import os
import urllib.request
from PIL import Image
```

```python
def pic2pdf(img_path,pdf_name):
    print(img_path)
    file_list = os.listdir(img_path)
    if len(file_list) == 0:
        return 
    pic_name = []
    im_list = []
    for x in file_list:
        if "jpg" in x or 'png' in x or 'jpeg' in x:
            pic_name.append(x)
    pic_name.sort(key=lambda x:int(x.split('.')[0]))
    new_pic = []

    for x in pic_name:
        if "jpg" in x:
            new_pic.append(x)

    for x in pic_name:
        if "png" in x:
            new_pic.append(x)

    print("hec", new_pic)

    try:
        im1 = Image.open(os.path.join(img_path,new_pic[0]))
        new_pic.pop(0)
        for i in new_pic:
            try:
                img = Image.open(os.path.join(img_path,i))
                if img.mode == "RGBA":
                    img = img.convert('RGB')
                    im_list.append(img)
                else:
                    im_list.append(img)
            except OSError:
                pass
    except IndexError:
        pass
    im1.save(os.path.join('H:\data\Spider\data\pdf\\',str(pdf_name)+'.pdf'), "PDF", resolution=100.0, save_all=True, append_images=im_list)
    del im_list
    del im1
    print("输出文件名称：", pdf_name)
```

```python
all_paths = [os.path.join('H:\data\Spider\data\details\\',_) for _ in os.listdir('H:\data\Spider\data\details')]
all_paths_name = [ os.path.basename(_) for _ in all_paths]
[pic2pdf(all_paths[index],all_paths_name[index]) for index,value in enumerate(all_paths) if index >= 100 & index <=160]
```
