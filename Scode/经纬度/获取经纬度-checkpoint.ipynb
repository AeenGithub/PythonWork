```python
import requests
import time 
import json
import xlwings as xw
import pandas as pd
import numpy as np
```

```python
path = r"D:\其他文件\供应商主数据.xls"
app = xw.App(visible=True, add_book=False)
```

```python
xb = app.books.open(path)
```

```python
rng = xb.sheets[1].range('a1:e3000').options(pd.DataFrame).value
rng.dropna(axis=0, how='all', inplace=True)
rng.columns
```

```python
name = rng['供应商名称'].tolist()
# place = rng['办公地址'].tolist()
```

```python
def getpois(key):   
    url_api = r'http://restapi.amap.com/v3/place/text'
    pois_all = [] 
    dic = {'key':'a116cd29b51b3e642af47a866dd35076',
           'keywords':key,
           'offset':'20',
           'output':'json',
           'extensions':'all',
           'page':1
          }
    try:
        req = requests.get(url_api,params=dic)
        json_one = json.loads(req.text)
        pois = json_one['pois']
        pois_all.append(pois)
    except Exception as e:
        print(e)
    return pois_all
```

```python
def getgps(pois):
    count = 0
    lat_lng = []
    lst_res = []
    while count < 1:
        lst_just = []
        name = pois[count]['name']
        lng, lat = pois[count]['location'].split(',')
        address = pois[count]['address']
        lst_just = [name, lng, lat, address]
        lst_res.append(lst_just)
        count = count + 1
    for one in lst_res:
        print(one[0], one[1], one[2], one[3], '\n')
        lat_lng.append(one[1])
        lat_lng.append(one[2])
    return lat_lng
```

```python
all_lat_lng = []
for i, key_one in enumerate(name):
    print(i, key_one)
    try:
        pois_ALL = getpois(key_one)
        for pois_one in pois_ALL:
            if bool(pois_one) is True:
                tup = [key_one, getgps(pois_one)]
                all_lat_lng.append(tup)
    except IndexError as e:
        print(e)
        continue
```
